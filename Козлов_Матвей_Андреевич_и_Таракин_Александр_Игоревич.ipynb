```python
# =====================================================
# ĞŸĞ¾Ğ»Ğ½Ğ¾ÑĞ²ÑĞ·Ğ½Ğ°Ñ Ğ½ĞµĞ¹Ñ€Ğ¾Ğ½Ğ½Ğ°Ñ ÑĞµÑ‚ÑŒ TensorFlow Ğ´Ğ»Ñ Ğ¾Ñ†ĞµĞ½ĞºĞ¸
# ÑÑ„Ñ„ĞµĞºÑ‚Ğ¸Ğ²Ğ½Ğ¾ÑÑ‚Ğ¸ Ğ¼Ğ°Ñ€ĞºĞµÑ‚Ğ¸Ğ½Ğ³Ğ¾Ğ²Ñ‹Ñ… ĞºĞ°Ğ¼Ğ¿Ğ°Ğ½Ğ¸Ğ¹ (Bank Marketing UCI)
# ĞŸĞ¾Ğ»Ğ½Ğ¾ÑÑ‚ÑŒÑ Ñ€Ğ°Ğ±Ğ¾Ñ‚Ğ¾ÑĞ¿Ğ¾ÑĞ¾Ğ±Ğ½Ñ‹Ğ¹ ĞºĞ¾Ğ´ Ğ´Ğ»Ñ Google Colab
# =====================================================

# Ğ£ÑÑ‚Ğ°Ğ½Ğ¾Ğ²ĞºĞ° Ğ½ĞµĞ¾Ğ±Ñ…Ğ¾Ğ´Ğ¸Ğ¼Ñ‹Ñ… Ğ±Ğ¸Ğ±Ğ»Ğ¸Ğ¾Ñ‚ĞµĞº
!pip install ucimlrepo tensorflow scikit-learn pandas numpy matplotlib seaborn -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from ucimlrepo import fetch_ucirepo
import warnings
warnings.filterwarnings('ignore')

print("TensorFlow version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))
print("="*60)

# =====================================================
# 1. Ğ—ĞĞ“Ğ Ğ£Ğ—ĞšĞ Ğ˜ ĞĞĞĞ›Ğ˜Ğ— Ğ”ĞĞ¢ĞĞ¡Ğ•Ğ¢Ğ BANK MARKETING UCI
# =====================================================
print("ğŸ“Š Ğ—ĞĞ“Ğ Ğ£Ğ—ĞšĞ Ğ”ĞĞ¢ĞĞ¡Ğ•Ğ¢Ğ BANK MARKETING UCI")
print("="*60)

# Ğ—Ğ°Ğ³Ñ€ÑƒĞ·ĞºĞ° Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ğ° Ğ½Ğ°Ğ¿Ñ€ÑĞ¼ÑƒÑ Ğ¸Ğ· UCI ML Repository
bank_marketing = fetch_ucirepo(id=222)  # UCI ID 222 [web:11][web:20]

# Ğ”Ğ°Ğ½Ğ½Ñ‹Ğµ
X = bank_marketing.data.features
y = bank_marketing.data.targets

print("Ğ Ğ°Ğ·Ğ¼ĞµÑ€ Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ğ°: X =", X.shape, "y =", y.shape)
print("\nĞŸĞµÑ€Ğ²Ñ‹Ğµ 5 ÑÑ‚Ñ€Ğ¾Ğº:")
print(X.head())
print("\nĞ¦ĞµĞ»ĞµĞ²Ğ°Ñ Ğ¿ĞµÑ€ĞµĞ¼ĞµĞ½Ğ½Ğ°Ñ (y):")
print(y.head())

# =====================================================
# 2. Ğ˜ĞĞ–Ğ•ĞĞ•Ğ Ğ˜Ğ¯ ĞŸĞ Ğ˜Ğ—ĞĞĞšĞĞ’ Ğ˜ ĞŸĞ Ğ•Ğ”ĞĞ‘Ğ ĞĞ‘ĞĞ¢ĞšĞ
# =====================================================
print("\nğŸ”§ ĞŸĞ Ğ•Ğ”ĞĞ‘Ğ ĞĞ‘ĞĞ¢ĞšĞ Ğ”ĞĞĞĞ«Ğ¥")
print("="*60)

# ĞĞ±ÑŠĞµĞ´Ğ¸Ğ½ÑĞµĞ¼ X Ğ¸ y Ğ´Ğ»Ñ ÑƒĞ´Ğ¾Ğ±ÑÑ‚Ğ²Ğ° Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ¸
df = pd.concat([X, y], axis=1)
print("ĞŸĞ¾Ğ»Ğ½Ñ‹Ğ¹ Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚:", df.shape)
print("\nĞ Ğ°ÑĞ¿Ñ€ĞµĞ´ĞµĞ»ĞµĞ½Ğ¸Ğµ ĞºĞ»Ğ°ÑÑĞ¾Ğ²:")
print(df['y'].value_counts(normalize=True))

# ĞšĞ¾Ğ´Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ Ñ†ĞµĞ»ĞµĞ²Ğ¾Ğ¹ Ğ¿ĞµÑ€ĞµĞ¼ĞµĞ½Ğ½Ğ¾Ğ¹ (yes/no -> 1/0)
le_y = LabelEncoder()
df['y'] = le_y.fit_transform(df['y'])
print("\nĞŸĞ¾ÑĞ»Ğµ ĞºĞ¾Ğ´Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ñ y:")
print(df['y'].value_counts(normalize=True))

# ĞĞ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ° ĞºĞ°Ñ‚ĞµĞ³Ğ¾Ñ€Ğ¸Ğ°Ğ»ÑŒĞ½Ñ‹Ñ… Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¾Ğ²
categorical_cols = df.select_dtypes(include=['object']).columns.drop('y', errors='ignore')
print("\nĞšĞ°Ñ‚ĞµĞ³Ğ¾Ñ€Ğ¸Ğ°Ğ»ÑŒĞ½Ñ‹Ğµ Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¸:", list(categorical_cols))

# One-Hot Encoding Ğ´Ğ»Ñ ĞºĞ°Ñ‚ĞµĞ³Ğ¾Ñ€Ğ¸Ğ°Ğ»ÑŒĞ½Ñ‹Ñ… Ğ¿ĞµÑ€ĞµĞ¼ĞµĞ½Ğ½Ñ‹Ñ…
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
print("ĞŸĞ¾ÑĞ»Ğµ one-hot encoding:", df_encoded.shape)

# Ğ Ğ°Ğ·Ğ´ĞµĞ»ĞµĞ½Ğ¸Ğµ Ğ¾Ğ±Ñ€Ğ°Ñ‚Ğ½Ğ¾ Ğ½Ğ° Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¸ Ğ¸ Ñ†ĞµĞ»ĞµĞ²ÑƒÑ Ğ¿ĞµÑ€ĞµĞ¼ĞµĞ½Ğ½ÑƒÑ
X_processed = df_encoded.drop('y', axis=1)
y_processed = df_encoded['y']

print("Ğ¤Ğ¸Ğ½Ğ°Ğ»ÑŒĞ½Ñ‹Ğµ Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¸:", X_processed.shape)
print("ĞšĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ ĞºĞ»Ğ°ÑÑĞ¾Ğ²: 0-No, 1-Yes")

# ĞĞ¾Ñ€Ğ¼Ğ°Ğ»Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ Ñ‡Ğ¸ÑĞ»Ğ¾Ğ²Ñ‹Ñ… Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¾Ğ²
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_processed)
X_scaled = pd.DataFrame(X_scaled, columns=X_processed.columns)

print("\nâœ… Ğ”Ğ°Ğ½Ğ½Ñ‹Ğµ Ğ¿Ğ¾Ğ´Ğ³Ğ¾Ñ‚Ğ¾Ğ²Ğ»ĞµĞ½Ñ‹ Ğ´Ğ»Ñ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ")
print("X_train shape:", X_scaled.shape, "y_train shape:", y_processed.shape)

# =====================================================
# 3. ĞŸĞĞ”Ğ“ĞĞ¢ĞĞ’ĞšĞ Ğ”ĞĞĞĞ«Ğ¥ Ğ”Ğ›Ğ¯ ĞĞ‘Ğ£Ğ§Ğ•ĞĞ˜Ğ¯
# =====================================================
# Ğ Ğ°Ğ·Ğ´ĞµĞ»ĞµĞ½Ğ¸Ğµ Ğ½Ğ° train/validation/test (70/15/15)
X_temp, X_test, y_temp, y_test = train_test_split(
    X_scaled, y_processed, test_size=0.15, random_state=42, stratify=y_processed
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1765, random_state=42, stratify=y_temp  # 15/85=0.1765
)

print(f"Train: {X_train.shape[0]} ({X_train.shape[0]/len(X_scaled)*100:.1f}%)")
print(f"Val:   {X_val.shape[0]} ({X_val.shape[0]/len(X_scaled)*100:.1f}%)")
print(f"Test:  {X_test.shape[0]} ({X_test.shape[0]/len(X_scaled)*100:.1f}%)")

# =====================================================
# 4. ĞŸĞĞ ĞĞœĞ•Ğ¢Ğ Ğ˜Ğ—ĞĞ¦Ğ˜Ğ¯ ĞœĞĞ”Ğ•Ğ›Ğ˜ (ĞŸĞĞ›ĞĞĞ¯ ĞĞĞ¡Ğ¢Ğ ĞĞ™ĞšĞ)
# =====================================================
print("\nğŸ§  ĞšĞĞĞ¤Ğ˜Ğ“Ğ£Ğ ĞĞ¦Ğ˜Ğ¯ ĞĞ•Ğ™Ğ ĞĞĞĞĞ™ Ğ¡Ğ•Ğ¢Ğ˜")
print("="*60)

# ĞŸĞ¾Ğ»Ğ½Ğ°Ñ Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
MODEL_CONFIG = {
    'input_dim': X_train.shape[1],
    'hidden_layers': [256, 128, 64],  # ĞŸĞ¾Ğ»Ğ½Ğ¾ÑĞ²ÑĞ·Ğ½Ñ‹Ğµ ÑĞ»Ğ¾Ğ¸
    'dropout_rate': 0.3,
    'learning_rate': 0.001,
    'batch_size': 128,
    'epochs': 50,
    'early_stopping_patience': 10,
    'optimizer': 'adam',
    'loss': 'binary_crossentropy',
    'metrics': ['accuracy', 'auc', 'precision', 'recall']
}

print("ĞšĞ¾Ğ½Ñ„Ğ¸Ğ³ÑƒÑ€Ğ°Ñ†Ğ¸Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸:")
for key, value in MODEL_CONFIG.items():
    print(f"  {key}: {value}")

# =====================================================
# 5. ĞŸĞĞ¡Ğ¢Ğ ĞĞ•ĞĞ˜Ğ• ĞŸĞĞ›ĞĞĞ¡Ğ’Ğ¯Ğ—ĞĞĞ™ ĞĞ•Ğ™Ğ ĞĞĞĞĞ™ Ğ¡Ğ•Ğ¢Ğ˜
# =====================================================
def create_model(config):
    """Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ğµ Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ğ¸Ğ·Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ½Ğ¾Ğ¹ Ğ¿Ğ¾Ğ»Ğ½Ğ¾ÑĞ²ÑĞ·Ğ½Ğ¾Ğ¹ ĞĞ¡"""
    model = keras.Sequential([
        layers.Input(shape=(config['input_dim'],)),

        # ĞŸĞµÑ€Ğ²Ñ‹Ğ¹ ÑĞºÑ€Ñ‹Ñ‚Ñ‹Ğ¹ ÑĞ»Ğ¾Ğ¹
        layers.Dense(config['hidden_layers'][0], activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(config['dropout_rate']),

        # Ğ”Ğ¾Ğ¿Ğ¾Ğ»Ğ½Ğ¸Ñ‚ĞµĞ»ÑŒĞ½Ñ‹Ğµ ÑĞºÑ€Ñ‹Ñ‚Ñ‹Ğµ ÑĞ»Ğ¾Ğ¸
        layers.Dense(config['hidden_layers'][1], activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(config['dropout_rate']),

        layers.Dense(config['hidden_layers'][2], activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(config['dropout_rate']),

        # Ğ’Ñ‹Ñ…Ğ¾Ğ´Ğ½Ğ¾Ğ¹ ÑĞ»Ğ¾Ğ¹ Ğ´Ğ»Ñ Ğ±Ğ¸Ğ½Ğ°Ñ€Ğ½Ğ¾Ğ¹ ĞºĞ»Ğ°ÑÑĞ¸Ñ„Ğ¸ĞºĞ°Ñ†Ğ¸Ğ¸
        layers.Dense(1, activation='sigmoid', name='output')
    ])

    return model

# Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
model = create_model(MODEL_CONFIG)
model.summary()

# =====================================================
# 6. ĞšĞĞœĞŸĞ˜Ğ›Ğ¯Ğ¦Ğ˜Ğ¯ Ğ˜ ĞĞ‘Ğ£Ğ§Ğ•ĞĞ˜Ğ• ĞœĞĞ”Ğ•Ğ›Ğ˜
# =====================================================
print("\nğŸš€ ĞĞ‘Ğ£Ğ§Ğ•ĞĞ˜Ğ• ĞœĞĞ”Ğ•Ğ›Ğ˜")
print("="*60)

# ĞšĞ¾Ğ¼Ğ¿Ğ¸Ğ»ÑÑ†Ğ¸Ñ
optimizer = tf.keras.optimizers.Adam(learning_rate=MODEL_CONFIG['learning_rate'])
model.compile(
    optimizer=optimizer,
    loss=MODEL_CONFIG['loss'],
    metrics=MODEL_CONFIG['metrics']
)

# Callbacks
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_auc', patience=MODEL_CONFIG['early_stopping_patience'],
        restore_best_weights=True, verbose=1
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.5, patience=5, min_lr=1e-7, verbose=1
    ),
    keras.callbacks.ModelCheckpoint(
        'best_marketing_model.h5', monitor='val_auc', save_best_only=True
    )
]

# ĞĞ±ÑƒÑ‡ĞµĞ½Ğ¸Ğµ
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=MODEL_CONFIG['epochs'],
    batch_size=MODEL_CONFIG['batch_size'],
    callbacks=callbacks,
    verbose=1
)

print("âœ… ĞĞ±ÑƒÑ‡ĞµĞ½Ğ¸Ğµ Ğ·Ğ°Ğ²ĞµÑ€ÑˆĞµĞ½Ğ¾!")

# =====================================================
# 7. ĞĞĞĞ›Ğ˜Ğ— Ğ Ğ•Ğ—Ğ£Ğ›Ğ¬Ğ¢ĞĞ¢ĞĞ’ ĞĞ‘Ğ£Ğ§Ğ•ĞĞ˜Ğ¯
# =====================================================
print("\nğŸ“ˆ ĞĞĞĞ›Ğ˜Ğ— Ğ Ğ•Ğ—Ğ£Ğ›Ğ¬Ğ¢ĞĞ¢ĞĞ’")
print("="*60)

# Ğ“Ñ€Ğ°Ñ„Ğ¸ĞºĞ¸ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Loss
axes[0,0].plot(history.history['loss'], label='Train Loss')
axes[0,0].plot(history.history['val_loss'], label='Val Loss')
axes[0,0].set_title('Model Loss')
axes[0,0].legend()

# Accuracy
axes[0,1].plot(history.history['accuracy'], label='Train Acc')
axes[0,1].plot(history.history['val_accuracy'], label='Val Acc')
axes[0,1].set_title('Model Accuracy')
axes[0,1].legend()

# AUC
axes[1,0].plot(history.history['auc'], label='Train AUC')
axes[1,0].plot(history.history['val_auc'], label='Val AUC')
axes[1,0].set_title('Model AUC')
axes[1,0].legend()

# Precision/Recall
axes[1,1].plot(history.history['precision'], label='Precision')
axes[1,1].plot(history.history['val_precision'], label='Val Precision')
axes[1,1].plot(history.history['recall'], label='Recall')
axes[1,1].plot(history.history['val_recall'], label='Val Recall')
axes[1,1].set_title('Precision & Recall')
axes[1,1].legend()

plt.tight_layout()
plt.show()

# =====================================================
# 8. ĞĞ¦Ğ•ĞĞšĞ ĞĞ Ğ¢Ğ•Ğ¡Ğ¢ĞĞ’ĞĞ™ Ğ’Ğ«Ğ‘ĞĞ ĞšĞ•
# =====================================================
print("\nğŸ¯ ĞĞ¦Ğ•ĞĞšĞ ĞĞ Ğ¢Ğ•Ğ¡Ğ¢ĞĞ’ĞĞ™ Ğ’Ğ«Ğ‘ĞĞ ĞšĞ•")
print("="*60)

# ĞŸÑ€ĞµĞ´ÑĞºĞ°Ğ·Ğ°Ğ½Ğ¸Ñ
y_pred_proba = model.predict(X_test, verbose=0).flatten()
y_pred = (y_pred_proba > 0.5).astype(int)

# ĞœĞµÑ‚Ñ€Ğ¸ĞºĞ¸
test_loss, test_accuracy, test_auc, test_precision, test_recall = model.evaluate(X_test, y_test, verbose=0)

print(f"Test Accuracy:  {test_accuracy:.4f}")
print(f"Test AUC-ROC:   {test_auc:.4f}")
print(f"Test Precision: {test_precision:.4f}")
print(f"Test Recall:    {test_recall:.4f}")
print(f"Test Loss:      {test_loss:.4f}")

# ĞŸĞ¾Ğ´Ñ€Ğ¾Ğ±Ğ½Ñ‹Ğ¹ Ğ¾Ñ‚Ñ‡ĞµÑ‚
print("\nğŸ“‹ CLASSIFICATION REPORT:")
print(classification_report(y_test, y_pred, target_names=['No Subscription', 'Yes Subscription']))

# ĞœĞ°Ñ‚Ñ€Ğ¸Ñ†Ğ° Ğ¾ÑˆĞ¸Ğ±Ğ¾Ğº
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])
plt.title('Confusion Matrix (Test Set)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# ROC ĞºÑ€Ğ¸Ğ²Ğ°Ñ
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc_score(y_test, y_pred_proba):.4f}')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (Test Set)')
plt.legend()
plt.grid(True)
plt.show()

# =====================================================
# 9. Ğ‘Ğ˜Ğ—ĞĞ•Ğ¡-Ğ˜ĞĞ¢Ğ•Ğ ĞŸĞ Ğ•Ğ¢ĞĞ¦Ğ˜Ğ¯ Ğ Ğ•Ğ—Ğ£Ğ›Ğ¬Ğ¢ĞĞ¢ĞĞ’
# =====================================================
print("\nğŸ’¼ Ğ‘Ğ˜Ğ—ĞĞ•Ğ¡-Ğ˜ĞĞ¢Ğ•Ğ ĞŸĞ Ğ•Ğ¢ĞĞ¦Ğ˜Ğ¯")
print("="*60)

# ROI Ğ°Ğ½Ğ°Ğ»Ğ¸Ğ· (Ğ¿Ñ€Ğ¸Ğ¼ĞµÑ€Ğ½Ğ°Ñ Ğ¾Ñ†ĞµĞ½ĞºĞ°)
total_clients = len(X_test)
predicted_yes = sum(y_pred)
actual_yes = sum(y_test)

cost_per_call = 2  # USD Ğ·Ğ° Ğ·Ğ²Ğ¾Ğ½Ğ¾Ğº
profit_per_subscription = 50  # USD Ğ¿Ñ€Ğ¸Ğ±Ñ‹Ğ»ÑŒ Ñ Ğ¿Ğ¾Ğ´Ğ¿Ğ¸ÑĞºĞ¸

# Ğ¡Ğ»ÑƒÑ‡Ğ°Ğ¹Ğ½Ñ‹Ğµ Ğ·Ğ²Ğ¾Ğ½ĞºĞ¸ (Ğ±Ğ°Ğ·Ğ»Ğ°Ğ¹Ğ½ ~11% ĞºĞ¾Ğ½Ğ²ĞµÑ€ÑĞ¸Ñ [web:17])
baseline_subscriptions = total_clients * 0.11
baseline_profit = baseline_subscriptions * profit_per_subscription - total_clients * cost_per_call

# ĞœĞ¾Ğ´ĞµĞ»ÑŒ (Ğ·Ğ²Ğ¾Ğ½ĞºĞ¸ Ñ‚Ğ¾Ğ»ÑŒĞºĞ¾ predicted_yes)
model_subscriptions = sum(y_pred & y_test)  # True positives
model_profit = model_subscriptions * profit_per_subscription - predicted_yes * cost_per_call

roi_improvement = ((model_profit - baseline_profit) / abs(baseline_profit)) * 100

print(f"Ğ’ÑĞµĞ³Ğ¾ ĞºĞ»Ğ¸ĞµĞ½Ñ‚Ğ¾Ğ² Ğ² Ñ‚ĞµÑÑ‚Ğµ: {total_clients:,}")
print(f"Ğ‘Ğ°Ğ·Ğ»Ğ°Ğ¹Ğ½ ĞºĞ¾Ğ½Ğ²ĞµÑ€ÑĞ¸Ñ (ÑĞ»ÑƒÑ‡Ğ°Ğ¹Ğ½Ğ¾): {baseline_subscriptions:.0f} Ğ¿Ğ¾Ğ´Ğ¿Ğ¸ÑĞ¾Ğº")
print(f"Ğ‘Ğ°Ğ·Ğ»Ğ°Ğ¹Ğ½ Ğ¿Ñ€Ğ¸Ğ±Ñ‹Ğ»ÑŒ: ${baseline_profit:.2f}")
print(f"ĞœĞ¾Ğ´ĞµĞ»ÑŒ Ğ¿Ñ€ĞµĞ´ÑĞºĞ°Ğ·Ğ°Ğ»Ğ°: {predicted_yes} ĞºĞ»Ğ¸ĞµĞ½Ñ‚Ğ¾Ğ²")
print(f"ĞœĞ¾Ğ´ĞµĞ»ÑŒ Ğ¿Ğ¾Ğ´Ñ‚Ğ²ĞµÑ€Ğ¶Ğ´ĞµĞ½Ğ¾: {model_subscriptions} Ğ¿Ğ¾Ğ´Ğ¿Ğ¸ÑĞ¾Ğº")
print(f"ĞŸÑ€Ğ¸Ğ±Ñ‹Ğ»ÑŒ Ñ Ğ¼Ğ¾Ğ´ĞµĞ»ÑŒÑ: ${model_profit:.2f}")
print(f"ğŸ“ˆ Ğ£Ğ»ÑƒÑ‡ÑˆĞµĞ½Ğ¸Ğµ ROI: +{roi_improvement:.1f}%")

print("\n" + "="*60)
print("âœ… ĞŸĞĞ›ĞĞ«Ğ™ ĞĞĞĞ›Ğ˜Ğ— Ğ—ĞĞ’Ğ•Ğ Ğ¨Ğ•Ğ!")
print("ĞœĞ¾Ğ´ĞµĞ»ÑŒ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ° ĞºĞ°Ğº 'best_marketing_model.h5'")
print("="*60)

# Ğ¡Ğ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ¸Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
model.save('bank_marketing_full_model.h5')
print("ĞœĞ¾Ğ´ĞµĞ»ÑŒ ÑĞ¾Ñ…Ñ€Ğ°Ğ½ĞµĞ½Ğ°: bank_marketing_full_model.h5")
```
